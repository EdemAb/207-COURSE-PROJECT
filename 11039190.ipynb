```python
#obtaining mean values to replace Na values 

mean_fixed_acidity = dataset['fixed acidity'].mean()
mean_volatile_acidity = dataset['volatile acidity'].mean()
mean_chlorides = dataset['chlorides'].mean()
mean_free_sulfur_dioxide = dataset['free sulfur dioxide'].mean()
mean_density = dataset['density'].mean()
mean_quality = dataset['quality'].mean()
```

```python
mean_fixed_acidity = mean_fixed_acidity.round(2)
mean_volatile_acidity = mean_volatile_acidity.round(2)
mean_chlorides = mean_chlorides.round(3)
mean_free_sulfur_dioxide = mean_free_sulfur_dioxide.round(1)
mean_density = mean_density.round(4)
mean_quality = mean_quality.round(1)
```

```python
dataset['fixed acidity'] = dataset['fixed acidity'].fillna(mean_fixed_acidity)
dataset['volatile acidity'] = dataset['volatile acidity'].fillna(mean_volatile_acidity)
dataset['chlorides'] = dataset['chlorides'].fillna(mean_chlorides)
dataset['free sulfur dioxide'] = dataset['free sulfur dioxide'].fillna(mean_free_sulfur_dioxide)
dataset['density'] = dataset['density'].fillna(mean_density)
dataset['quality'] = dataset['quality'].fillna(mean_quality)
```

```python
mode_fixed_acidity = dataset['fixed acidity'].mode()
mode_residual_sugar = dataset['residual sugar'].mode()
mode_volatile_acidity = dataset['volatile acidity'].mode()
mode_citric_acid = dataset['citric acid'].mode()
mode_chlorides = dataset['chlorides'].mode()
mode_free_sulfur_dioxide = dataset['free sulfur dioxide'].mode()
mode_total_sulfur_dioxide = dataset['total sulfur dioxide'].mode()
mode_pH = dataset['pH'].mode()
mode_alcohol = dataset['alcohol'].mode()
mode_sulphates = dataset['sulphates'].mode()
mode_density = dataset['density'].mode()
mode_quality = dataset['quality'].mode()
```

```python
print(f'fixed acidity mode: {mode_fixed_acidity}') 
print(f'volatile acidity mode: {mode_volatile_acidity }')
print(f'citric acid mode: {mode_citric_acid }')
print(f'residual sugar mode: {mode_residual_sugar}')
print(f'alcohol mode: {mode_alcohol}')
print(f'chlorides mode: {mode_chlorides}') 
print(f'free sulfur dioxide mode: {mode_free_sulfur_dioxide}')
print(f'free total dioxide mode: {mode_total_sulfur_dioxide}')
print(f'pH mode: {mode_pH}')
print(f'sulphates mode: {mode_sulphates}')
print(f'density mode: {mode_density}')
print(f'quality mode: {mode_quality}')
```

```python
plt.figure(figsize=(19, 19))
plt.suptitle('HISTOGRAM OF VARIOUS DATA COLUMNS', fontsize=25)
for i in range(dataset.shape[1]):
    plt.subplot(6, 3, i + 1)
    sub_graph = plt.gca()
    sub_graph.set_title(dataset.columns.values[i])

    vals = np.size(dataset.iloc[:, i].unique())
    
    plt.hist(dataset.iloc[:, i], bins=vals, color='green')
```

```python
FIG, axes = plt.subplots(3, 3, figsize=[15,12])
plt.subplots_adjust(wspace=0.2 , hspace=1)
dataInfo.loc['mean'].plot(ylabel='mean distribution', rot=45, ax=axes[0, 0], color='orange').set_title('MEAN')
dataInfo.loc['std'].plot(ylabel='standard deviation', rot=45, ax=axes[0, 1]).set_title('STANDARD DEVIATION')
dataInfo.loc['min'].plot(ylabel='minimum distribution', rot=45, ax=axes[0, 2], color='yellow').set_title('MINIMUM')
dataInfo.loc['25%'].plot(ylabel="25%_distribution", rot=45, ax=axes[1, 0], color='pink').set_title('LOWER QUARTILE')
dataInfo.loc['50%'].plot(ylabel='50%_distribution', rot=45, ax=axes[1, 1], color='gray').set_title('MEDIAN')
dataInfo.loc['75%'].plot(ylabel='75%_distribuiton', rot=45, ax=axes[1, 2], color='skyblue').set_title('UPPER QUARTILE')
dataInfo.loc['max'].plot(ylabel='maximum distribution', rot=45, ax=axes[2, 1], color='red').set_title('MAXIMUM')

```
